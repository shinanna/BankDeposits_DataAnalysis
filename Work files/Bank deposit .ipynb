```python
## Encoding 
```

```python
month_labels = {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6, 'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12}
data['month'] = data['month'].apply(lambda x: month_labels[x])

data
```

```python
poutcome_labels = {'unknown': 0, 'failure': 1,'other': 2, 'success': 3}
data['poutcome'] = data['poutcome'].apply(lambda x: poutcome_labels[x])
data
```

```python
contact_labels = {'unknown': 0, 'cellular': 1,'telephone': 2,}
data['contact'] = data['contact'].apply(lambda x: contact_labels[x])
data
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data["Job_encode"]=encoder.fit_transform(data["job"])
data["Education_encode"]=encoder.fit_transform(data["education"])
data["Marital_encode"]=encoder.fit_transform(data["marital"])
data["Default_encode"]=encoder.fit_transform(data["default"])
data["Housing_encode"]=encoder.fit_transform(data["housing"])
data["Loan_encode"]=encoder.fit_transform(data["loan"])
data["Subscription _encode"]=encoder.fit_transform(data["Subscription to deposit (target)"])
data=data.drop(columns=["education", "marital", "job","default","housing","loan","Subscription to deposit (target)"])
data
```
